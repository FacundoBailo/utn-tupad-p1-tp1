```python
#Ejercicio 1

notas = []

for i in range(10):
    nota = float(input("Ingrese la nota del estudiante"))
    notas.append(nota)

promedio = sum(notas) / len(notas)
max_nota = max(notas)
min_nota = min(notas)

print(f"La lista completa de notas es: {notas}, El promedio de las notas en la lista nos da: { promedio }, la nota mas alta es: { max_nota } y la nota mas baja es: { min_nota}")

```

```python
#Ejercicio 2

productos = []

for i in range(5):
    producto = input(f"Ingrese el prodcuto {i+1}")
    productos.append(producto)

lista_ordenada = sorted(productos)

print(f"Lista de productos ordenados alfaveticamente: {lista_ordenada}")

eliminar = input("Ingrese el nombre del producto que desea elminar de la lista")

while eliminar not in lista_ordenada:
    eliminar = input(f"El nombre no existe, ingrese un nombre de la siguiente lista: {lista_ordenada}")

lista_ordenada.remove(eliminar)

print(f"Lista actualizada: { lista_ordenada }")

```

```python
#Ejercicio 5

estudiantes = []
agregar_eliminar = 0

for i in range(8):
    estudante = input(f"Ingrese el nombre del Estudiante {i+1}")
    estudiantes.append(estudante)

while agregar_eliminar !=3:
    agregar_eliminar = int(input("Desea agregar o elminar algun estudiante? (Elija el numero correspondiente)" \
    "1. Agregar " \
    "2. Eliminar " \
    "3. Salir "))

    if agregar_eliminar == 1:
        estudante = input("Ingrese el nombre del nuevo estudiante")
        estudiantes.append( estudante )
    elif agregar_eliminar == 2:
        estudante = input("Ingrese el nombre del estudiante que quiere elminar de la lista")
        
        while estudante not in estudiantes:
            estudante = input("El nombre ingresado no se encuentra en la lista, intente nuevamente con otro nombre")
        
        estudiantes.remove(estudante)
    
print(estudiantes)

```

```python
#Ejercicio 7 

dias = []
amplitudes = []

for i in range(7):

    for j in range(2):

        temperatura = []

        minima = float(input("Ingrese la temperatura minima del dia"))
        temperatura.append(minima)


        maxima = float(input("Ingrese la temperatura maxima del dia"))
        temperatura.append(maxima)
        
    dias.append(temperatura)

promedio_minimas = sum(dia[0] for dia in dias) / 7
promedio_maximas = sum(dia[1] for dia in dias) / 75

for dia in dias:
    amplitud = dia[1] - dia[0]
    amplitudes.append(amplitud)

max_amplitud = max(amplitudes)

dia_max_amplitud = amplitudes.index(max_amplitud) + 1


print(f"Promedio de minimas: {promedio_minimas}ºC, promedio de maximas: {promedio_maximas}ºC, dia con maxima amplitud: {dia_max_amplitud} con una amplitud de: {max_amplitud}ºC")

```

```python
#Ejercicio 8

notas = []
for i in range(5):
    fila = []

    for j in range(3):
        nota = float(input(f"  Nota de la materia {j+1}: "))
        fila.append(nota)
    notas.append(fila)

for i, fila in enumerate(notas, start=1):
    prom_est = sum(fila) / len(fila)
    print(f"Promedio del estudiante {i}: {prom_est:.2f}")

materias = len(notas[0])
for j in range(materias):
    prom_mat = sum(notas[i][j] for i in range(len(notas))) / len(notas)
    print(f"Promedio de la materia {j+1}: {prom_mat:.2f}")
```

```python
#Ejercicio 9

tablero = [["-" for _ in range(3)] for _ in range(3)]

def mostrar_tablero():
    for fila in tablero:
        print(" ".join(fila))
    print() 

def pedir_jugada(jugador):
    while True:
        try:
            fila = int(input(f"Jugador {jugador} - fila (0-2): "))
            col  = int(input(f"Jugador {jugador} - columna (0-2): "))

            if 0 <= fila < 3 and 0 <= col < 3 and tablero[fila][col] == "-":
                return fila, col
            else:
                print("Posición inválida o ya ocupada. Intenta de nuevo.")
        except ValueError:
            print("Debes ingresar números enteros del 0 al 2.")

turnos = 0
while turnos < 6:  
    jugador = "X" if turnos % 2 == 0 else "O"
    fila, col = pedir_jugada(jugador)
    tablero[fila][col] = jugador
    mostrar_tablero()
    turnos += 1
```

```python
#Ejercicio 10

ventas = []

for i in range(4):   
    fila = []
    print(f"\nProducto {i+1}")
    for j in range(7):   
        venta = float(input(f"  Ventas del día {j+1}: "))
        fila.append(venta)
    ventas.append(fila)


for i, fila in enumerate(ventas, start=1):
    total_prod = sum(fila)
    print(f"Total vendido del producto {i}: {total_prod}")


totales_por_dia = [sum(ventas[f][c] for f in range(len(ventas)))
                   for c in range(len(ventas[0]))]
mayor_ventas_dia = max(totales_por_dia)
dia_mayor = totales_por_dia.index(mayor_ventas_dia) + 1  

print(f"Día con mayores ventas totales: Día {dia_mayor} (Total: {mayor_ventas_dia})")


totales_productos = [sum(fila) for fila in ventas]
max_prod = max(totales_productos)
producto_top = totales_productos.index(max_prod) + 1

print(f"Producto más vendido en la semana: Producto {producto_top} (Total: {max_prod})")

```
